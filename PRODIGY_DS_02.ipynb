```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path=('/content/drive/MyDrive/train.csv')
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df.info()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.histplot(df['Age'], kde=True, ax=axes[0, 0])
sns.histplot(df['Fare'], kde=True, ax=axes[0, 1])
sns.countplot(x='Pclass', data=df, ax=axes[1, 0])
sns.countplot(x='Survived', data=df, ax=axes[1, 1])
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df, ci=False, palette='rainbow')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.barplot(x='Sex_male', y='Survived', data=df, ci=False, palette='rocket')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='plasma')
```
